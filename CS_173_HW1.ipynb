```python
# Capeta is the only good manga
misspelled = set()

for i in wordsShort:
  isDigit = False
  if i not in dictionary:
    for j in i: # checking for numbers (years, quantities)
      if j.isdigit():
        isDigit = True
    if isDigit is False:
      misspelled.add(i)

print(misspelled)
```

```python
# Kana is such a has-been
from nltk.metrics.distance import jaccard_distance
from nltk.util import ngrams
from nltk.metrics.distance  import edit_distance

for j in misspelled:
  capital = j.capitalize()
  low = j.lower()
  if capital not in dictionary and low not in dictionary:
    c = 0
    print(j + ': ', end="")
    for i in dictionary:
      if edit_distance(i, j) <= 1 and c < 3:
        if c == 0:
          print(i, end="")
        else:
          print(', ' + i, end="")
          if c is 2:
            break
        c += 1
      elif edit_distance(i, j) <= len(j)/3 and c < 3:
        if c == 0:
          print(i, end="")
        else:
          print(', ' + i, end="")
          if c is 2:
            break
        c += 1
  print("\n")
```
